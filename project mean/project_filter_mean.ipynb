```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import imageio
import cv2
from skimage import data, color
from skimage.transform import rescale, resize
```

```python
tinggi = 250
lebar = 250
```

```python
#menambahkan noise pada gambar
def filterr(img, nilai=0.005):
  n, m = img.shape
  noisy = np.copy(img)

  for i in range(0, n):
    for j in range(0, m):
      num = np.random.randint(10000, size=1)
      if (num < nilai*10000):
        noisy[i][j] = np.random.randint(2, size=1)*255

  return noisy
```

```python
# baca gambar uji dan lakukan pra-pemrosesan
# ubah ukuran, skala abu-abu, derau
pic = imageio.imread("pic.jpg")

image_resized = resize(pic, (tinggi, lebar), anti_aliasing=True)
grayimage = color.rgb2gray(image_resized) * 255
img1 = grayimage.astype(np.uint8)
img1 = filterr(img1)
```

```python
# bandingkan gambar asli dengan gambar yang sudah diproses sebelumnya

plt.figure(figsize=(10, 10))
plt.subplot(121); plt.imshow(pic, cmap='gray', vmin=0, vmax=255); plt.axis('off')
plt.subplot(122); plt.imshow(img1, cmap='gray', vmin=0, vmax=255); plt.axis('off')
```
